```python
import matplotlib.pyplot as plt
import numpy as np
salaries = [55, 62, 65, 58, 70, 55, 60, 68, 63, 55, 62, 150]
bins = [50, 60, 70, 80, 90, 100, 110, 120, 130, 140, 150]
counts, bin_edges = np.histogram(salaries, bins=bins)
plt.bar(bin_edges[:-1], counts, width=np.diff(bin_edges))
plt.title("Distribution of Employee Salaries")
plt.xlabel("Salary (in thousands of dollars)")
plt.ylabel("Frequency")

plt.show()
```
