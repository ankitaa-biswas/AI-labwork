```python

import matplotlib.pyplot as plt
import pandas as pd


data = {
    'Subject': ['Mathematics', 'Physics', 'Chemistry', 'Mathematics', 'Physics'],
    'Author': ['R.D. Sharma', 'H.C. Verma', 'O.P. Tandon', 'S.K. Goyal', 'Resnick'],
    'Quantity': [15, 12, 8, 10, 5]
}

df=pd.DataFrame(data)
subject_totals = df.groupby('Subject')['Quantity'].sum()

subject_totals.plot(kind='bar')

plt.title('Total Books per Subject')
plt.xlabel('Subject')
plt.ylabel('Number of Books')

plt.xticks(rotation=360)

plt.show()
```
